```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
RESULT = "run_03"
```

```python
df = pd.read_csv(f"../results/{RESULT}.csv")
```

```python
PARTICLE_COLORS = [
    "#FA4656", "#2C73D6", "#00D75B", "#FEF058", "#FFAA4C", "#A241B2"]
```

```python
# Plot configuration
params = {"axes.labelsize": 6.0,
          "axes.titlesize": 6.0,
          "figure.dpi": 300,
          "figure.figsize": [3.0, 2.0],
          "figure.facecolor": 'white',
          "font.size": 5.0,
          "xtick.top": 'on',
          "xtick.labelsize": 5.0,
          "ytick.labelsize": 5.0,
          "ytick.right": 'on',
          "xtick.major.width": 0.75,
          "xtick.major.size": 2.0,
          "ytick.major.width": 0.75,
          "ytick.major.size": 2.0,
          "xtick.direction": "in",
          "ytick.direction": "in",
          "axes.grid": False,
          "legend.framealpha": 0.0,
          "lines.linewidth": 1.0,}
plt.rcParams.update(params)
```

```python
fig, ax = plt.subplots()

ax.set_xlabel(r"x", weight=600)
ax.set_ylabel(r"y", weight=600)

for i in range(3):
    try:
        ax.plot(df[f"xPosition{i}"], df[f"yPosition{i}"], lw=0.75,
                color=PARTICLE_COLORS[i])
        ax.scatter(df[f"xPosition{i}"].iloc[0], df[f"yPosition{i}"].iloc[0],
                   color=PARTICLE_COLORS[i], s=5, marker='s')
        ax.scatter(df[f"xPosition{i}"].iloc[-1], df[f"yPosition{i}"].iloc[-1],
                   color=PARTICLE_COLORS[i], s=5, marker='o')
    except KeyError:
        pass

plt.suptitle("Trayectorias de las partículas", x=0.1, color="#868E96",
             weight=800, ha="left", size=8)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.set_xlabel(r"Tiempo", weight=600)
ax.set_ylabel(r"Energía", weight=600)

ax.plot(df["Time"], df["Energy"], lw=0.75,
        color=PARTICLE_COLORS[0], label="Mecánica")
ax.plot(df["Time"], df["Potential"], lw=0.75,
        color=PARTICLE_COLORS[1], label="Potencial")
ax.plot(df["Time"], df["KineticEnergy"], lw=0.75,
        color=PARTICLE_COLORS[2], label="Cinética")

leg = ax.legend(loc="best", framealpha=1)
leg.get_frame().set_linewidth(0.0)

plt.suptitle("Evolución de la energía", x=0.1, color="#868E96",
             weight=800, ha="left", size=8)

plt.show()
```
